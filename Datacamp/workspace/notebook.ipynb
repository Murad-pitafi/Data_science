```python
# Filter for rows where family_members is less than 1000 
# and region is Pacific
fam_lt_1k_pac = homelessness[(homelessness['family_members'] < 1000) & (homelessness['region'].isin(['Pacific']))]

# See the result
print(fam_lt_1k_pac)
```
